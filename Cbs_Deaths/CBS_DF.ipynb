```python
read_file = pd.read_excel (r'27_20_355t1.xls')
read_file.to_csv (r'cbs.csv', index = None, header=True)
```

```python
    
# Imported cbs.csv
cbs = pd.read_csv(r'cbs.csv', encoding='utf-8')

# Deleted columns Unnamed: 2
cbs.drop(['Unnamed: 2'], axis=1, inplace=True)

# Deleted columns Unnamed: 3
cbs.drop(['Unnamed: 3'], axis=1, inplace=True)

# Deleted columns Unnamed: 4
cbs.drop(['Unnamed: 4'], axis=1, inplace=True)

# Deleted columns Unnamed: 5
cbs.drop(['Unnamed: 5'], axis=1, inplace=True)

# Deleted columns Unnamed: 6
cbs.drop(['Unnamed: 6'], axis=1, inplace=True)

# Renamed columns killed
cbs.rename(columns={'Unnamed: 1': 'killed'}, inplace=True)

# Renamed columns year
cbs.rename(columns={'תרשים 1 - הרוגים בתאונות דרכים 2019-1949, מספרים מוחלטים': 'year'}, inplace=True)

# Dropped duplicates in cbs
cbs = cbs.drop_duplicates(keep=False)
```

```python
#if run twice it will delete a row again ***run once only***
# Dropped first corrupted row
cbs=cbs.drop(cbs.index[0])
cbs
```

```python
# https://www.gov.il/he/departments/news/20201231_summary
# adding killed in 2020
cbs2 = {'year': '2020', 'killed': 302}
cbs = cbs.append(cbs2, ignore_index = True)
```

```python
# https://www.gov.il/he/departments/news/20211231_summary
# adding killed in 2021
cbs2 = {'year': '2021', 'killed': 361}
cbs = cbs.append(cbs2, ignore_index = True)
```
