```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
SloveniaDf=pd.read_csv("../CleanDataFrames/Slovenia.csv")
```

```python
SloveniaDf
```

```python
SloveniaDf.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
lr = linear_model.LinearRegression()
x = SloveniaDf[['Total inland transport infrastructure investment',
'Rail infrastructure maintenance',
'Road motor vehicles per one million units of current USD GDP',
'Density of road (km per one hundred sq. km)',
'Total railway enterprises - Number at 31.12']]
y = SloveniaDf['Killed']
lr.fit(X=x, y=y)
```

Linear regression scores:

```python
print("Slope:",lr.coef_)
print("Intercept:",lr.intercept_)
print("R2:",lr.score(x,y))
print("R2:",r2_score(y,lr.predict(x.values)))
```
