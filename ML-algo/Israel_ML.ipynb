```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
IsraelDF=pd.read_csv("../CleanDataFrames/Israel.csv")
```

```python
IsraelDF
```

```python
IsraelDF.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
lr = linear_model.LinearRegression()
x = IsraelDF[['Road motor vehicles per one million units of current USD GDP',
'Share of household expenditure for purchase of vehicles in total household expenditure for transport']]
y = IsraelDF['Killed']
lr.fit(X=x, y=y)
```

Linear regression scores:

```python
print("Slope:",lr.coef_)
print("Intercept:",lr.intercept_)
print("R2:",lr.score(x,y))
print("R2:",r2_score(y,lr.predict(x.values)))
```
