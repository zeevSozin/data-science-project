```python
import requests
import pandas as pd
import time
import json
```

```python
base_url="https://data.gov.il/api/3/action/datastore_search?"
```

```python
ACCIDENT_TAZ_id="57c5aef9-70f9-4b71-82fa-52304cfbd031"
```

```python
resource_id=ACCIDENT_TAZ_id
```

```python
def get_record(base_url,resource_id,limit):
    responce=requests.get(base_url+"resource_id="+resource_id+"&limit="+limit)
    time.sleep(3)
    if (responce.status_code==200):
        json_raw_data=json.loads(responce.text)
        json_data=json_raw_data['result']
        dict_1=json_data["records"]
        df=pd.DataFrame(dict_1)
    else:
        print("no responce!!!")
        
    return df
    
```

```python
limit="300"
data_300=get_record(base_url,resource_id,limit)
```
